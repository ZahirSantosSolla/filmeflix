```python
import pandas as pd
import numpy as np

from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
# 2. Geração do Dataset

#O conjunto de dados foi gerado artificialmente conforme especificado no enunciado.
#Cada usuário possui um gênero favorito e tende a avaliar melhor filmes desse gênero.
```

```python
def generate_recommendation_data(
    n_users=500,
    n_movies=50
):

    np.random.seed(42)

    generos = [
        'Ação',
        'Drama',
        'Comédia',
        'Sci-Fi',
        'Terror'
    ]

    filmes = [
        f'Filme_{i}'
        for i in range(n_movies)
    ]

    filme_genero = {
        filme: np.random.choice(generos)
        for filme in filmes
    }

    data=[]

    for user_id in range(1,n_users+1):

        genero_favorito=np.random.choice(generos)

        qtd=np.random.randint(10,20)

        vistos=np.random.choice(
            filmes,
            qtd,
            replace=False
        )

        for filme in vistos:

            if filme_genero[filme] == genero_favorito:
                nota=np.random.randint(4,6)
            else:
                nota=np.random.randint(1,5)

            data.append([
                user_id,
                filme,
                nota,
                filme_genero[filme]
            ])

    df = pd.DataFrame(
        data,
        columns=[
            'ID_Usuario',
            'ID_Filme',
            'Nota',
            'Genero'
        ]
    )

    df.to_csv(
        'filme_flix_ratings.csv',
        index=False
    )

    return df


df = generate_recommendation_data()

df.head()
```

```python
# 4. Construção da Matriz Usuário-Item
```

```python
matriz = df.pivot_table(
    index='ID_Usuario',
    columns='ID_Filme',
    values='Nota'
)

matriz = matriz.fillna(0)

matriz.head()
```

```python
# 5. Treinamento do Modelo KNN
```

```python
modelo = NearestNeighbors(
    metric='cosine',
    algorithm='brute'
)

modelo.fit(matriz)
```

```python
# 6. Sistema de Recomendação
```

```python
def recomendar_filmes(
    usuario_id,
    matriz,
    modelo,
    n_recomendacoes=5
):

    indice = usuario_id - 1

    distancias, indices = modelo.kneighbors(
        [matriz.iloc[indice]],
        n_neighbors=6
    )

    usuarios_similares = indices.flatten()[1:]

    usuario_alvo = matriz.iloc[indice]

    recomendados = set()

    for vizinho in usuarios_similares:

        notas_vizinho = matriz.iloc[vizinho]

        filmes_bons = notas_vizinho[
            (notas_vizinho >=4) &
            (usuario_alvo==0)
        ].index

        recomendados.update(
            filmes_bons
        )

    return list(
        recomendados
    )[:n_recomendacoes]


print(
    recomendar_filmes(
        1,
        matriz,
        modelo
    )
)
```

```python
# 7. A "Mágica" da Similaridade

#Se dois usuários possuem padrões semelhantes de avaliações, o modelo entende que suas preferências são parecidas.

#Assim, filmes bem avaliados por um usuário similar podem ser recomendados ao outro.
```

```python
# 8. Avaliação do Modelo
```

```python
real = df['Nota']

pred = df['Nota'].mean()

rmse = sqrt(
    mean_squared_error(
        real,
        [pred]*len(real)
    )
)

print("RMSE:",rmse)
```

```python
#O RMSE representa o erro médio das notas previstas. Quanto menor esse valor, mais próximas as previsões estão das notas reais dos usuários.
```
